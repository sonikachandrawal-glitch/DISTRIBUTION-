ANS.4 STANDARIZATION (Z-SCORE)

     STANDARDIZATION IS THE PROCESS OF CONVERTING RAW DATA VALUE INTO Z-SCORES , WHICH SHOW HOW FAR A VALUE IS FORM THE MEAN , MEASURED IN STANDARD DEVIATIONS.

     Z- SCORE TELL US

      HOW UNUSUAL OR TYPICAL A DATA POINT IS

      Z- SCORE FORMULAS - X= ORIGINAL DATA VALUE
                          U = MEAN OF THE DATASET
                          STANDARD DEVIATION OF DATASET

      STANDARDIZATION TRANSFORMS A DATASET

      AFTER STANDARDIZATION

      THE NEW MEAN BECOMES ZERO

      THE NEW STANDARD DEVIATION BECOMES 1

      DATA IS RESCALED BUT RELATIVE POSITIONS REMAINS THE SAME

      STANDARIZATION IS IMPORTANT IN DATA ANALYSIS BECAUSE

      COMPARISON ACROSS DIFFRENT SCALES

      IDENTIFYING THE OUTLIERS

      REQUIRED FOR MANY ALGORITHMS

      INTERPRETING THE PROBABILITIES
      

Q.5 What is Kurtosis and their type?

ANS.5 KURTOSIS IS A STATISTICAL MEASURE THAT DESCRIBES THE S
      SHAPE OF A DISTRIBUTION , SPEACIALLY THE TAILEDENESS OR
      PEAKEDNESS OF THE DATA.

      KURTOSIS TELL US

      HOW HEAVY OR LIGHT THE TAILS OF A DISTRIBUTION ARE

      THE TENDENCY OF A DATASET TO PRODUCE OUTLIERS

      TYPES OF KURTOSIS

      1. MESOKURTIC

         KURTOSIS = 3 (EXCESS KURTOSIS = 0 )

         SHAPE SIMILAR TO NOMAL DISTRIBUTION

         MODERATE TAILS AND PEAK
        
          eg HEIGHTS OF PEOPLE

      2. LEPTOKURTIC

         KURTOSIS > 3 (EXCESS KURTOSIS = 0)

         HIGH PEAK AND HEAVY TAILS

         MORE EXTREME VALUES

      3. PLATYKURTIC

        KURTOSIS < 3 ( EXCESS KURTOSIS < 0)

        FLAT PEAK AND LIGHT TAILS

        FEWER OUTLIERS

Q.8 How does log transformation help in stabilizing variance and making data more normally distributed?

ANS.8 LOG TRANSFORMATION IS A COMMON DATA- PREPROCESSING
      TECHNIQUE USED TO REDUCE SKEWNESS, STABILIZE VARIIANCE,
      AND MAKE DATA MORE NORMALLY DISTRIBUTED , ESPECIALLY WHEN VALUES SPAN SEVERAL ORDERS OF MAGNITUDE.

      LOG TRANSFORMATION REPLACES EACH DATA VALUE X WITH ITS ALGORITHMS , TYPICALLY

      Y = LOG (X) (OR LOG10X,InX)

      LOG TRANSFORMATION HELPS

      1.STABILIES VARIANCE (REDUCE HETROSCEDASTICITY)

       LOG TRANSFORMATION COMPRESS LARGE VALUES MORE
       THAN SMALL ONES , REDUCING UNEQUAL SPREAD .

       eg RAW DATA = VARIABILITY INCREASES AT HIGHER VALUES
          LOG DATA = VARIABILITY BECOMES MORE EVEN

      2.REDUCES RIGHT SKEWNESS

        MANY REAL-WORLD DATASETS ARE RIGHT-SKEWED
        (LONG TAIL ON THE RIGHT).

        TAKING LOGS PULLS IN EXTREME HIGH VALUES
        SHORTENING THE RIGHT TAIL.

        THIS DISTRIBUTION BECOMES MORE SYMMETRIC , CLOSER
        TO NORMAL

      3.HANDLES OUTLIERS

       LARGE OUTLIERS HAVE INFLUENCE AFTER LOG TRANSFORMATION

      4.IMPROVES MODEL ASSUMPTIONS

        LOG TRANSFORMATION HELPS SATISFY ASSUMPTIONS OF

         NORMALITY , CONSTANT VARIENCE , LINEARNITY(IN REGRESSION)
